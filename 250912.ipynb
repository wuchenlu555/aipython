NumPy는 배열 조작, 수학 연산, 통계 분석 등 다양한 작업을 위한 풍부한 메소드를 제공합니다. 주요 카테고리는 다음과 같습니다.

*   **배열 생성 및 조작:** `np.array()`, `np.zeros()`, `np.ones()`, `np.reshape()`, `np.transpose()`, `np.concatenate()` 등
*   **수학 연산:** `np.add()`, `np.subtract()`, `np.multiply()`, `np.divide()`, `np.sqrt()`, `np.sin()`, `np.cos()` 등
*   **통계 함수:** `np.mean()`, `np.median()`, `np.std()`, `np.sum()`, `np.max()`, `np.min()` 등
*   **선형 대수:** `np.dot()`, `np.linalg.inv()`, `np.linalg.det()` 등
*   **난수 생성:** `np.random.rand()`, `np.random.randn()`, `np.random.randint()` 등

이 외에도 많은 메소드가 있으며, 필요에 따라 [NumPy 공식 문서](https://numpy.org/doc/stable/reference/index.html)를 참고하시면 더욱 자세한 정보를 얻으실 수 있습니다.

```python
import matplotlib.pyplot as plt
# 한글 폰트 설정 (한글 깨짐 방지)
plt.rcParams['font.family'] = 'Malgun Gothic' # Windows
plt.rcParams['axes.unicode_minus'] = False
# 간단한 그래프 그리기
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.figure(figsize=(8, 5))
plt.plot(x, y, 'bo-', label='데이터')
plt.title('간단한 그래프 예제')
plt.xlabel('X 값')
plt.ylabel('Y 값')
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
공부시간 = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)
```

```python
시험점수 = np.array([50, 55, 65, 70, 75, 85, 90, 95])
```

```python
for i in range(len(공부시간)):
 print(f"공부시간: {공부시간[i][0]}시간 → 점수: {시험점수[i]}점")
```

```python
model = LinearRegression()
model.fit(공부시간, 시험점수)
```

```python
새로운_시간 = np.array([[9]]) # 6.5시간 공부했을 때
예측_점수 = model.predict(새로운_시간)
```

```python
예측_점수
```

```python
print(f"\n예측 결과:")
print(f"6.5시간 공부하면 예상 점수: {예측_점수[0]:.1f}점")
```
