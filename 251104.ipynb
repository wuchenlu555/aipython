```python
import pandas as pd
import numpy as np

# Create a sample DataFrame for demonstration
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Heidi'],
    'Age': [24, 27, 22, 32, 29, 45, 30, 25],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'New York', 'Los Angeles', 'Chicago'],
    'Salary': [50000, 60000, 45000, 75000, 62000, 90000, 58000, 48000],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'HR'],
    'Experience': [2, 4, 1, 8, 5, 15, 6, 3],
    'StartDate': pd.to_datetime(['2022-01-15', '2020-05-20', '2023-03-01', '2016-11-10', '2019-07-22', '2009-02-01', '2018-09-05', '2021-04-12'])
}
df = pd.DataFrame(data)

df.loc[2, 'Salary'] = np.nan # Introduce a missing value
df.loc[5, 'Department'] = np.nan # Introduce another missing value

display(df)
```

### 1. Data Loading and Saving

These methods are used to read data from various file formats into a DataFrame and to save a DataFrame to a file.

### 2. Data Viewing and Inspection

These methods help you get a quick overview of your DataFrame's structure, content, and basic statistics.

```python
# .head(n): Displays the first n rows (default is 5)
print("df.head():")
display(df.head())

# .tail(n): Displays the last n rows (default is 5)
print("\ndf.tail(3):")
display(df.tail(3))

# .info(): Provides a concise summary of the DataFrame, including data types and non-null values
print("\ndf.info():")
df.info()

# .describe(): Generates descriptive statistics of numerical columns
print("\ndf.describe():")
display(df.describe())

# .shape: Returns a tuple representing the dimensionality of the DataFrame (rows, columns)
print(f"\ndf.shape: {df.shape}")

# .columns: Returns the column labels of the DataFrame
print(f"\ndf.columns: {df.columns.tolist()}")

# .index: Returns the index (row labels) of the DataFrame
print(f"\ndf.index: {df.index.tolist()}")

# .values: Returns a NumPy representation of the DataFrame
print("\ndf.values (first 3 rows):")
display(df.values[:3])
```

### 3. Data Selection and Filtering

These methods are crucial for selecting specific rows and columns from your DataFrame.

```python
# Selecting a single column
print("df['Name'] (Series):")
display(df['Name'].head())

# Selecting multiple columns
print("\ndf[['Name', 'Salary']] (DataFrame):")
display(df[['Name', 'Salary']].head())

# .loc[]: Label-based indexing (selects by row and column labels)
print("\ndf.loc[0, 'Name']:")
print(df.loc[0, 'Name'])

print("\ndf.loc[1:3, ['Name', 'Age']]:")
display(df.loc[1:3, ['Name', 'Age']])

# .iloc[]: Integer-location based indexing (selects by row and column integer positions)
print("\ndf.iloc[0, 0] (first row, first column):")
print(df.iloc[0, 0])

print("\ndf.iloc[1:4, 0:2] (rows 1-3, columns 0-1):")
display(df.iloc[1:4, 0:2])

# Boolean indexing: Filtering rows based on a condition
print("\nEmployees with Age > 28:")
display(df[df['Age'] > 28])

# Combined conditions
print("\nEmployees in New York AND Salary > 50000:")
display(df[(df['City'] == 'New York') & (df['Salary'] > 50000)])
```

### 4. Data Cleaning and Manipulation

These methods help in handling missing values, renaming columns, and applying functions to data.

```python
# .drop(): Remove rows or columns
# Dropping a column
print("df after dropping 'Experience' column:")
display(df.drop(columns=['Experience']).head())

# Dropping rows (e.g., by index, or based on condition)
print("\ndf after dropping row with index 0:")
display(df.drop(index=0).head())

# .fillna(): Fill missing values
print("\ndf after filling NaN salaries with mean:")
display(df['Salary'].fillna(df['Salary'].mean()).head())

# Fill missing 'Department' with 'Unknown'
df_filled = df.copy()
df_filled['Department'] = df_filled['Department'].fillna('Unknown')
print("\ndf after filling NaN departments with 'Unknown':")
display(df_filled)

# .dropna(): Remove rows/columns with missing values
print("\ndf after dropping rows with any NaN values:")
display(df.dropna())

# .rename(): Rename columns or index labels
print("\ndf after renaming 'Age' to 'Years_Old':")
display(df.rename(columns={'Age': 'Years_Old'}).head())

# .apply(): Apply a function along an axis of the DataFrame
print("\nNew column 'Salary_USD' calculated using .apply():")
df['Salary_USD'] = df['Salary'].apply(lambda x: f"${x:,.2f}" if pd.notna(x) else np.nan)
display(df[['Name', 'Salary', 'Salary_USD']].head())

# .map(): Map values from one Series to another
rating_map = {'HR': 'Good', 'IT': 'Excellent', 'Finance': 'Very Good', 'Unknown': 'N/A'}
df_filled['Dept_Rating'] = df_filled['Department'].map(rating_map)
print("\ndf with 'Dept_Rating' column using .map():")
display(df_filled[['Department', 'Dept_Rating']].head())
```

### 5. Data Aggregation and Grouping

These methods are used to summarize data, often by grouping rows based on one or more criteria.

```python
# .groupby(): Group DataFrame using a mapper or by a Series of columns
print("Average Salary by Department:")
display(df.groupby('Department')['Salary'].mean())

# Group by multiple columns and aggregate multiple metrics
print("\nAggregated data by City and Department:")
display(df.groupby(['City', 'Department']).agg(
    Avg_Salary=('Salary', 'mean'),
    Min_Age=('Age', 'min'),
    Total_Employees=('Name', 'count')
))

# .value_counts(): Return a Series containing counts of unique values
print("\nValue counts for 'City':")
display(df['City'].value_counts())

print("\nValue counts for 'Department' (with dropna=False to include NaNs):")
display(df['Department'].value_counts(dropna=False))
```

### 6. Data Merging and Concatenation

These methods combine DataFrames in various ways, similar to SQL joins.

```python
# Create another DataFrame for demonstration
data2 = {
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'Population': [8.4, 4.0, 2.7, 2.3, 1.7]
}
df_cities = pd.DataFrame(data2)

print("Original df:")
display(df.head(2))
print("df_cities:")
display(df_cities)

# .merge(): Combines DataFrames based on common columns or indices
# Inner merge (default): only rows where the 'City' is present in both DataFrames
print("\nInner merge of df and df_cities on 'City':")
display(df.merge(df_cities, on='City', how='inner').head())

# Left merge: keeps all rows from the left DataFrame
print("\nLeft merge of df and df_cities on 'City':")
display(df.merge(df_cities, on='City', how='left'))

# .concat(): Stacks DataFrames vertically or horizontally
# Create another DataFrame with similar columns for vertical concatenation
data_new_employees = {
    'Name': ['Mia', 'Noah'],
    'Age': [26, 31],
    'City': ['Miami', 'Boston'],
    'Salary': [55000, 70000],
    'Department': ['IT', 'HR'],
    'Experience': [3, 7],
    'StartDate': pd.to_datetime(['2022-09-01', '2017-03-15'])
}
df_new = pd.DataFrame(data_new_employees)

print("\nConcatenated df and df_new (vertical, default):")
display(pd.concat([df, df_new], ignore_index=True))
```

### 7. Data Sorting

These methods arrange DataFrame rows by values in one or more columns or by index.

```python
# .sort_values(): Sort by the values along either axis
print("df sorted by 'Age' (ascending, default):")
display(df.sort_values(by='Age').head())

print("\ndf sorted by 'Salary' (descending):")
display(df.sort_values(by='Salary', ascending=False).head())

# Sort by multiple columns
print("\ndf sorted by 'Department' then by 'Salary' (descending):")
display(df.sort_values(by=['Department', 'Salary'], ascending=[True, False]))

# .sort_index(): Sort by index labels along either axis
df_indexed = df.set_index('Name')
print("\ndf_indexed sorted by index (Name) in descending order:")
display(df_indexed.sort_index(ascending=False).head())
```
